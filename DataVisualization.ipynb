```python
import pandas as pd
import numpy as np
import yfinance as yf
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.dates as mdates
import seaborn as sn
```

```python
df = pd.read_csv("finalizedDataset.csv")
stonks = pd.read_csv("stockPrices.csv")
benchmarks = pd.read_csv("benchmarkPrices.csv")
logReturnsStock = pd.read_csv("logReturnsStock.csv")
logReturnsBenchmark = pd.read_csv("logReturnsBenchmark.csv")
logReturnsStock['^GSPC'] = logReturnsBenchmark['^GSPC']
logReturnsStock['^TNX'] = logReturnsBenchmark['^TNX']
```

```python
stonks.set_index('Date', inplace = True)
stonks.index = pd.to_datetime(stonks.index).date
```

```python
plt.figure(figsize=(12,9))
ax = stonks[['AAPL','GOOG','TSLA','KO','OXY','BAC']].plot()
ax.xaxis.set_major_locator(mdates.DayLocator(interval=24))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.set_title('Graph of Individual Stocks')
plt.gcf().autofmt_xdate()
plt.show()
```

```python
benchmarks.set_index('Date', inplace = True)
benchmarks.index = pd.to_datetime(benchmarks.index).date
benchmarks['BAC'] = stonks['BAC']
benchmarks['OXY'] = stonks['OXY']
benchmarks['AAPL'] = stonks['AAPL']
benchmarks['GOOG'] = stonks['GOOG']
benchmarks['TSLA'] = stonks['TSLA']
benchmarks['KO'] = stonks['KO']
sn.heatmap(benchmarks.corr(), annot=True)
plt.show()
```
